```python
#@title { run: "auto", display-mode: "form" }
#@markdown <center><h3>Configure Zotify</h3></center><br>
#@markdown <center><i><b>Note:</b> Please run this cell once, after that, changes will be saved automatically every time you modify the values.</i><br><br>
import json
from typing import Any

# Editable config values

ROOT_PATH = "/content/drive/MyDrive/Zotify Music/" #@param {type:"string"}
ROOT_PODCAST_PATH = "/content/drive/MyDrive/Zotify Podcasts/" #@param {type:"string"}
DOWNLOAD_FORMAT = "ogg" #@param ["aac", "fdk_aac", "m4a", "mp3", "ogg", "opus", "vorbis"]
DOWNLOAD_QUALITY = "auto" #@param ["auto", "normal", "high", "very_high"]
BULK_WAIT_TIME = 5 #@param {type:"slider", min:1, max:120, step:1}
RETRY_ATTEMPTS = 5 #@param {type:"number"}
OVERRIDE_AUTO_WAIT = False #@param {type:"boolean"}
SKIP_EXISTING = True #@param {type:"boolean"}
SKIP_PREVIOUSLY_DOWNLOADED = False #@param {type:"boolean"}
DOWNLOAD_LYRICS = False #@param {type:"boolean"}
DOWNLOAD_REAL_TIME = False #@param {type:"boolean"}
SPLIT_ALBUM_DISCS = False #@param {type:"boolean"}
MD_SAVE_GENRES = True #@param {type:"boolean"}
MD_ALLGENRES = False #@param {type:"boolean"}
MD_GENREDELIMITER = ";" #@param {type:"string"}
LANGUAGE = "en" #@param {type:"string"}
CREDENTIALS_LOCATION = "/content/drive/MyDrive/credentials.json" #@param {type:"string"}

# Read-only default config values
SAVE_CREDENTIALS = True
SONG_ARCHIVE = ".song_archive"
CHUNK_SIZE = 50000
TRANSCODE_BITRATE = 'auto'
OUTPUT = ""
PRINT_SPLASH = False
PRINT_SKIPS = True
PRINT_DOWNLOAD_PROGRESS = True
PRINT_ERRORS = True
PRINT_DOWNLOADS = True
PRINT_API_ERRORS = False
PRINT_PROGRESS_INFO = True
PRINT_WARNINGS = True
TEMP_DOWNLOAD_DIR = ''

config_fp = './config.json'


CONFIG_VALUES = {
    'SAVE_CREDENTIALS':           { 'option': SAVE_CREDENTIALS},
    'CREDENTIALS_LOCATION':       { 'option': CREDENTIALS_LOCATION},
    'OUTPUT':                     { 'option': OUTPUT},
    'SONG_ARCHIVE':               { 'option': SONG_ARCHIVE},
    'ROOT_PATH':                  { 'option': ROOT_PATH},
    'ROOT_PODCAST_PATH':          { 'option': ROOT_PODCAST_PATH},
    'SPLIT_ALBUM_DISCS':          { 'option': SPLIT_ALBUM_DISCS},
    'DOWNLOAD_LYRICS':            { 'option': DOWNLOAD_LYRICS},   
    'MD_SAVE_GENRES':             { 'option': MD_SAVE_GENRES},
    'MD_ALLGENRES':               { 'option': MD_ALLGENRES},
    'MD_GENREDELIMITER':          { 'option': MD_GENREDELIMITER},
    'DOWNLOAD_FORMAT':            { 'option': DOWNLOAD_FORMAT},
    'DOWNLOAD_QUALITY':           { 'option': DOWNLOAD_QUALITY},
    'TRANSCODE_BITRATE':          { 'option': TRANSCODE_BITRATE},
    'SKIP_EXISTING':              { 'option': SKIP_EXISTING},
    'SKIP_PREVIOUSLY_DOWNLOADED': { 'option': SKIP_PREVIOUSLY_DOWNLOADED},
    'RETRY_ATTEMPTS':             { 'option': RETRY_ATTEMPTS},
    'BULK_WAIT_TIME':             { 'option': BULK_WAIT_TIME},
    'OVERRIDE_AUTO_WAIT':         { 'option': OVERRIDE_AUTO_WAIT},
    'CHUNK_SIZE':                 { 'option': CHUNK_SIZE},
    'DOWNLOAD_REAL_TIME':         { 'option': DOWNLOAD_REAL_TIME},
    'LANGUAGE':                   { 'option': LANGUAGE},
    'PRINT_SPLASH':               { 'option': PRINT_SPLASH},
    'PRINT_SKIPS':                { 'option': PRINT_SKIPS},
    'PRINT_DOWNLOAD_PROGRESS':    { 'option': PRINT_DOWNLOAD_PROGRESS},
    'PRINT_ERRORS':               { 'option': PRINT_ERRORS},
    'PRINT_DOWNLOADS':            { 'option': PRINT_DOWNLOADS},
    'PRINT_API_ERRORS':           { 'option': PRINT_API_ERRORS},
    'PRINT_PROGRESS_INFO':        { 'option': PRINT_PROGRESS_INFO},
    'PRINT_WARNINGS':             { 'option': PRINT_WARNINGS},
    'TEMP_DOWNLOAD_DIR':          { 'option': TEMP_DOWNLOAD_DIR}
}


class Config:
    Values = {}

    @classmethod
    def get_json(cls) -> Any:
        r = {}
        for key in CONFIG_VALUES:
            r[key] = CONFIG_VALUES[key]['option']
        return r

    @classmethod
    def write(cls) -> None:
        with open(config_fp, 'w', encoding='utf-8') as config_file:
            json.dump(cls.get_json(), config_file, indent=4)


Config.write()
print('Settings applied!')
```
